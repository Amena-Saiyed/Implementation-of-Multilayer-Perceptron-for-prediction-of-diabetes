### Using Perceptron algorithm to predict  diabetes

Step 1 : Import necessary librariers

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from IPython.display import display
%matplotlib inline
import requests
import numpy as np
```

Step 2: Load the pre-processed data

```python
from sklearn.datasets import load_svmlight_file
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from io import BytesIO

# URL of the dataset (pre-processed diabetes data)
diabetes_dataset_url = 'https://www.csie.ntu.edu.tw/~cjlin/libsvmtools/datasets/binary/diabetes'

# Downloading the dataset
dataset = requests.get(diabetes_dataset_url)

# Loading the dataset from the downloaded content
X, y = load_svmlight_file(BytesIO(dataset.content))

# Converting the sparse matrix to a dense array for further processing
X_dense = X.toarray()
```

Step 3 : Implementation of Multilayer Perceptron Algorithm

```python
# Split the dataset into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X_dense, y, test_size=0.2, random_state=42)

# Build the Multilayer Perceptron (MLP) Model
# using scikit-learn's MLPClassifier

mlp = MLPClassifier(hidden_layer_sizes=(10, 10), max_iter=1000, random_state=42)

# Train the MLP model
mlp.fit(X_train, y_train)

# Make predictions on the test set
y_pred = mlp.predict(X_test)
```

Step 4 : Model Evaluation

```python
# Evaluate the Model

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

# Print evaluation metrics
print(f"\nAccuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1 Score: {f1}")
```

```python
#  Plot the Training Loss Curve

loss_values = mlp.loss_curve_

# Plot the loss curve
plt.figure(figsize=(5, 5))
plt.plot(loss_values, label='Training Loss')
plt.title('Training Loss Curve for MLP')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(loc='upper right')
plt.grid(False)
plt.show()
```

```python
# implementing the whole architecture with adam and sgd

import requests
from io import BytesIO
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_svmlight_file
import matplotlib.pyplot as plt

# URL of the dataset (pre-processed diabetes data)
diabetes_dataset_url = 'https://www.csie.ntu.edu.tw/~cjlin/libsvmtools/datasets/binary/diabetes'

# Downloading the dataset
response = requests.get(diabetes_dataset_url)

# Loading the dataset from the downloaded content
X, y = load_svmlight_file(BytesIO(response.content))

# Convert the sparse matrix to a dense array for further processing
X_dense = X.toarray()

# Preprocess the target labels to be between 0 and 1
y = (y + 1) / 2

# Split dataset into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X_dense, y, test_size=0.3, random_state=42)

# Normalize the dataset (important for neural network training)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Convert to PyTorch tensors
X_train_tensor = torch.tensor(X_train, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train, dtype=torch.float32).view(-1, 1)  # Reshaping for binary classification
X_test_tensor = torch.tensor(X_test, dtype=torch.float32)
y_test_tensor = torch.tensor(y_test, dtype=torch.float32).view(-1, 1)

# Define the model architecture
class MLP(nn.Module):
    def __init__(self, input_dim):
        super(MLP, self).__init__()
        self.fc1 = nn.Linear(input_dim, 10)  # Hidden layer with 10 neurons
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(10, 1)  # Output layer with 1 neuron for binary classification
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        x = self.fc1(x)
        x = self.relu(x)
        x = self.fc2(x)
        x = self.sigmoid(x)
        return x

# Training function
def train_model(optimizer_name, num_epochs=200, lr=0.01):
    model = MLP(input_dim)
    criterion = nn.BCELoss()  # Binary Cross-Entropy Loss
    if optimizer_name == 'Adam':
        optimizer = optim.Adam(model.parameters(), lr=lr)
    elif optimizer_name == 'SGD':
        optimizer = optim.SGD(model.parameters(), lr=lr)
    else:
        raise ValueError("Unsupported optimizer")

    train_losses = []  # List to store training losses

    for epoch in range(num_epochs):
        model.train()  # Set the model to training mode

        # Forward pass
        outputs = model(X_train_tensor)
        loss = criterion(outputs, y_train_tensor)

        # Backward pass and optimization
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        # Store the loss value
        train_losses.append(loss.item())

        # Print training progress
        if (epoch + 1) % 20 == 0:
            train_accuracy = calculate_accuracy(y_train_tensor, outputs)
            print(f'Optimizer: {optimizer_name} | Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}, Accuracy: {train_accuracy.item():.4f}')

    # Evaluation on the test set
    model.eval()  # Set the model to evaluation mode
    with torch.no_grad():
        test_outputs = model(X_test_tensor)
        test_loss = criterion(test_outputs, y_test_tensor)
        test_accuracy = calculate_accuracy(y_test_tensor, test_outputs)

    return train_losses, test_loss.item(), test_accuracy.item()

# Function to calculate accuracy
def calculate_accuracy(y_true, y_pred):
    y_pred_labels = y_pred.round()  # Convert probabilities to binary labels (0 or 1)
    correct = (y_pred_labels == y_true).float()  # Calculate the number of correct predictions
    accuracy = correct.sum() / len(correct)  # Calculate the accuracy
    return accuracy

# Initialize parameters
input_dim = X_train.shape[1]
num_epochs = 200
lr = 0.01

# Train models with different optimizers and collect results
train_losses_adam, test_loss_adam, test_accuracy_adam = train_model('Adam', num_epochs, lr)
train_losses_sgd, test_loss_sgd, test_accuracy_sgd = train_model('SGD', num_epochs, lr)

# Plotting the training losses
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(range(1, num_epochs + 1), train_losses_adam, marker='o', label='Adam')
plt.plot(range(1, num_epochs + 1), train_losses_sgd, marker='x', label='SGD')
plt.title('Training Loss vs. Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

# Print test results
print(f"Adam Optimizer: Test Loss = {test_loss_adam:.4f}, Test Accuracy = {test_accuracy_adam:.4f}")
print(f"SGD Optimizer: Test Loss = {test_loss_sgd:.4f}, Test Accuracy = {test_accuracy_sgd:.4f}")

plt.subplot(1, 2, 2)
plt.bar(['Adam', 'SGD'], [test_accuracy_adam, test_accuracy_sgd], color=['blue', 'orange'])
plt.title('Test Accuracy Comparison')
plt.ylabel('Accuracy')
plt.ylim(0, 1)
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import requests
from io import BytesIO
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_svmlight_file
import matplotlib.pyplot as plt

# URL of the dataset (pre-processed diabetes data)
diabetes_dataset_url = 'https://www.csie.ntu.edu.tw/~cjlin/libsvmtools/datasets/binary/diabetes'

# Downloading the dataset
response = requests.get(diabetes_dataset_url)

# Check if the request was successful
if response.status_code == 200:
    print("Dataset downloaded successfully.")
else:
    raise Exception(f"Failed to download the dataset. Status code: {response.status_code}")

# Loading the dataset from the downloaded content
X, y = load_svmlight_file(BytesIO(response.content))

# Convert the sparse matrix to a dense array for further processing
X_dense = X.toarray()

# Preprocess the target labels to be between 0 and 1
y = (y + 1) / 2

# Split dataset into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X_dense, y, test_size=0.3, random_state=42)

# Normalize the dataset (important for neural network training)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Convert to PyTorch tensors
X_train_tensor = torch.tensor(X_train, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train, dtype=torch.float32).view(-1, 1)  # Reshaping for binary classification
X_test_tensor = torch.tensor(X_test, dtype=torch.float32)
y_test_tensor = torch.tensor(y_test, dtype=torch.float32).view(-1, 1)

# Define the model architecture
class MLP(nn.Module):
    def __init__(self, input_dim):
        super(MLP, self).__init__()
        self.fc1 = nn.Linear(input_dim, 10)  # Hidden layer with 10 neurons
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(10, 1)  # Output layer with 1 neuron for binary classification
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        x = self.fc1(x)
        x = self.relu(x)
        x = self.fc2(x)
        x = self.sigmoid(x)
        return x

# Training function
def train_model(optimizer_name, num_epochs=200, lr=0.01):
    model = MLP(input_dim)
    criterion = nn.BCELoss()  # Binary Cross-Entropy Loss
    if optimizer_name == 'Adam':
        optimizer = optim.Adam(model.parameters(), lr=lr)
    elif optimizer_name == 'SGD':
        optimizer = optim.SGD(model.parameters(), lr=lr)
    elif optimizer_name == 'Adagrad':
        optimizer = optim.Adagrad(model.parameters(), lr=lr)
    else:
        raise ValueError("Unsupported optimizer")

    train_losses = []  # List to store training losses

    for epoch in range(num_epochs):
        model.train()  # Set the model to training mode

        # Forward pass
        outputs = model(X_train_tensor)
        loss = criterion(outputs, y_train_tensor)

        # Backward pass and optimization
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        # Store the loss value
        train_losses.append(loss.item())

        # Print training progress
        if (epoch + 1) % 20 == 0:
            train_accuracy = calculate_accuracy(y_train_tensor, outputs)
            print(f'Optimizer: {optimizer_name} | Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}, Accuracy: {train_accuracy.item():.4f}')

    # Evaluation on the test set
    model.eval()  # Set the model to evaluation mode
    with torch.no_grad():
        test_outputs = model(X_test_tensor)
        test_loss = criterion(test_outputs, y_test_tensor)
        test_accuracy = calculate_accuracy(y_test_tensor, test_outputs)

    return train_losses, test_loss.item(), test_accuracy.item()

# Function to calculate accuracy
def calculate_accuracy(y_true, y_pred):
    y_pred_labels = y_pred.round()  # Convert probabilities to binary labels (0 or 1)
    correct = (y_pred_labels == y_true).float()  # Calculate the number of correct predictions
    accuracy = correct.sum() / len(correct)  # Calculate the accuracy
    return accuracy

# Initialize parameters
input_dim = X_train.shape[1]
num_epochs = 200
lr = 0.01

# Train models with different optimizers and collect results
train_losses_adam, test_loss_adam, test_accuracy_adam = train_model('Adam', num_epochs, lr)
train_losses_sgd, test_loss_sgd, test_accuracy_sgd = train_model('SGD', num_epochs, lr)
train_losses_adagrad, test_loss_adagrad, test_accuracy_adagrad = train_model('Adagrad', num_epochs, lr)

# Plotting the training losses
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(range(1, num_epochs + 1), train_losses_adam, marker='o', label='Adam')
plt.plot(range(1, num_epochs + 1), train_losses_sgd, marker='x', label='SGD')
plt.plot(range(1, num_epochs + 1), train_losses_adagrad, marker='^', label='Adagrad')
plt.title('Training Loss vs. Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(False)

# Print test results
print(f"Adam Optimizer: Test Loss = {test_loss_adam:.4f}, Test Accuracy = {test_accuracy_adam:.4f}")
print(f"SGD Optimizer: Test Loss = {test_loss_sgd:.4f}, Test Accuracy = {test_accuracy_sgd:.4f}")
print(f"Adagrad Optimizer: Test Loss = {test_loss_adagrad:.4f}, Test Accuracy = {test_accuracy_adagrad:.4f}")

plt.subplot(1, 2, 2)
plt.bar(['Adam', 'SGD', 'Adagrad'], [test_accuracy_adam, test_accuracy_sgd, test_accuracy_adagrad], color=['blue', 'orange', 'green'])
plt.title('Test Accuracy Comparison')
plt.ylabel('Accuracy')
plt.ylim(0, 1)
plt.grid(False)

plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score

# Function to calculate F1 score
def calculate_f1_score(y_true, y_pred):
    y_pred_labels = y_pred.round().numpy()  # Convert probabilities to binary labels (0 or 1)
    y_true = y_true.numpy()  # Convert tensor to NumPy array
    precision = precision_score(y_true, y_pred_labels)
    recall = recall_score(y_true, y_pred_labels)
    f1 = f1_score(y_true, y_pred_labels)
    return precision, recall, f1
```
